Nomor 4

Update dataframe df dengan menambahkan hasil perhitungan soal nomor 2, beri nama kolom BMI.

```python
#Fungsi Menghitung Berat Badan
def hitung_bmi (berat,tinggi):
  bmi=berat/(tinggi/100)**2
  return bmi

# Menghitung dan Menambahkan Kolom BMI
bmi_list = []
for index, row in data.iterrows():
    berat = row['Berat Badan (kg)']
    tinggi = row['Tinggi Badan (cm)']
    bmi = hitung_bmi(berat, tinggi)
    bmi_list.append(bmi)

# Menambahkan kolom BMI ke DataFrame
data['BMI'] = bmi_list

# Menampilkan DataFrame
data
```

Buatlah status BMI untuk setiap siswa dari kategori berikut, lalu simpan data statusnya dengan nama kolom Status BMI.

Normal: 18.5-24.9

Overweight: 25.0-29.9

Obesity: 30.0-34.9

Underweight: < 18.5



```python
#Fungsi untuk mengkategorikan BMI
def Status_BMI(BMI) :
  if bmi < 18.5:
        return "Underweight"
  elif 18.5 <= bmi <= 24.9:
        return "Normal"
  elif 25.0 <= bmi <= 29.9:
        return "Overweight"
  elif 30.0 <= bmi <= 34.9:
        return "Obesity"
  else:
        return "Severe Obesity"

# Mengkategorikan BMI
bmi_status = []
for index, row in data.iterrows():
    bmi = row['BMI']
    Status_bmi = Status_BMI(bmi)
    bmi_status.append(Status_bmi)

# Menambahkan kolom BMI ke DataFrame
data['Status BMI'] = bmi_status

# Menampilkan DataFrame
data
```

Soal 2

Perusahaan Investasi Jamin-Sugih Co kami memberikan penawaran investasi terbaik dengan pembagian keuntungan tetap sebesar 25% setiap tahun.

Buatlah program Python yang dapat menerima input dari pengguna dalam bentuk:
Modal Investasi (bilangan bulat antara 0-1.000.000.000)
Jangka Waktu Investasi (bilangan bulat antara 1-25 dalam tahun)
Selain itu, program akan menampilkan error message bahwa modal atau tenure tidak sesuai ketentuan.

Program akan menampilkan informasi akumulasi investasi setiap tahun, menggunakan iterasi untuk setiap tahun.



```python
def hitung_investasi(modal, tenure):
    # Cek apakah modal dan tenure sesuai ketentuan
    if not (0 < modal <= 1000000000):
        print("Error: Modal investasi harus dalam rentang 1 hingga 1.000.000.000")
        return
    if not (0 < tenure <= 25):
        print("Error: Jangka waktu investasi harus dalam rentang 1 hingga 25 tahun")
        return

    # Inisialisasi variabel untuk menyimpan keuntungan dan total modal
    total_modal = modal

    # Iterasi untuk setiap tahun
    for tahun in range(1, tenure + 1):
        # Hitung keuntungan tahunan
        keuntungan_tahunan = 0.25 * total_modal
        # Tambahkan keuntungan tahunan ke total modal
        total_modal += keuntungan_tahunan

        # Tampilkan informasi investasi untuk tahun tersebut
        print(f"Year {tahun}, Your Total Investment {total_modal:,.2f}")

# Input modal investasi
modal_input = int(input("Masukkan modal investasi (antara 1 hingga 1000000000): "))

# Input jangka waktu investasi
tenure_input = int(input("Masukkan jangka waktu investasi (antara 1 hingga 25 tahun): "))

# Hitung dan tampilkan informasi investasi
hitung_investasi(modal_input, tenure_input)

```
